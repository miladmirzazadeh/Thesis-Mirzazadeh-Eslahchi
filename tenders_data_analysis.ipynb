```python
import pandas as pd 
it_tenders_2019= pd.read_csv("Opentender/data-it-csv/data-it-2019.csv", sep=';')
```

```python
bid_price_available = it_tenders_2019[it_tenders_2019["bid_price_EUR"].notna()]
lot_estimated_price_available = it_tenders_2019[it_tenders_2019["lot_estimatedPrice_EUR"].notna()]
winning = it_tenders_2019[it_tenders_2019["bid_isWinning"]=="yes"]
```

```python
import numpy as np 
winning_with_bids = it_tenders_2019[np.logical_and(it_tenders_2019.bid_price_EUR.notna(), it_tenders_2019.bid_isWinning == "yes")]
```

```python
winning_with_estimated_price = it_tenders_2019[np.logical_and(it_tenders_2019.lot_estimatedPrice_EUR.notna(), it_tenders_2019.bid_isWinning == "yes")]
```

```python
rows_with_both_bid_and_estimated = it_tenders_2019[np.logical_and(it_tenders_2019.lot_estimatedPrice_EUR.notna(), it_tenders_2019.bid_price_EUR.notna())]
```

```python
rent = rows_with_both_bid_and_estimated.lot_estimatedPrice_EUR - rows_with_both_bid_and_estimated.bid_price_EUR
```

```python
print("mean of rents: ", np.mean(rent))
print("sigma of rents: " , np.std(rent))
```

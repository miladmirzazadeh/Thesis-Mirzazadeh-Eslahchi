# STEP 1: PREPARE OPENTENDER (The Winners)

```python
import pandas as pd
import numpy as np

# 1. SETUP: Create an independent copy
df_ot = it_tenders_2019.copy()

# 2. FILTER: Keep only Winners with a valid name
# We filter for 'yes' string and ensure bidder_name is not empty
df_winners = df_ot[
    (df_ot['bid_isWinning'] == 'yes') & 
    (df_ot['bidder_name'].notna())
].copy()

# 3. DATE LOGIC: Extract the "Join Year"
# Priority: 'tender_awardDecisionDate' -> Fallback: 'tender_year'
df_winners['award_date'] = pd.to_datetime(df_winners['tender_awardDecisionDate'], errors='coerce')
df_winners['join_year'] = df_winners['award_date'].dt.year.fillna(df_winners['tender_year'])

# Drop rows where year is still missing (cannot be merged)
df_winners = df_winners.dropna(subset=['join_year'])
df_winners['join_year'] = df_winners['join_year'].astype(int)

# 4. ID CLEANING: Standardize the Bidder ID
# Fill missing bidder_id with the alternative bodyId if available
df_winners['bidder_id_filled'] = df_winners['bidder_id'].fillna(df_winners['bidder_bodyId_id'])

# Remove all non-numeric characters (e.g., "IT 12345" becomes "12345")
df_winners['clean_id'] = df_winners['bidder_id_filled'].astype(str).str.replace(r'\D', '', regex=True)

print(f"Step 1 Complete. Winning Tenders ready: {len(df_winners)}")
```
